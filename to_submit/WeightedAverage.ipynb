```python
# Phase 1 Result: 0.82268 - Weighted average example code
import pandas as pd

files = ['/content/drive/MyDrive/to_submit/Results/Phase1/kcbert_001.csv', 
         '/content/drive/MyDrive/to_submit/Results/Phase1/kcbert_00.csv', 
         '/content/drive/MyDrive/to_submit/Results/Phase1/electra.csv', 
         '/content/drive/MyDrive/to_submit/Results/Phase1/klue-roberta-large.csv']
ratio = [3, 4, 1, 5]
sum = 0
for i in ratio:
  sum += i
sz = len(files)
df = pd.read_csv(files[0])
df = df * ratio[0]
files.pop(0)
ratio.pop(0)
for i in range(len(files)):
  tdf = pd.read_csv(files[i])
  tdf = tdf * ratio[i]
  df = df.add(tdf, fill_value=0)

df = df / sum
df["idx"] = df["idx"].astype('int')

df.to_csv('/content/drive/MyDrive/to_submit/Results/Phase1/weighted_average_result.csv', index = False)
```
